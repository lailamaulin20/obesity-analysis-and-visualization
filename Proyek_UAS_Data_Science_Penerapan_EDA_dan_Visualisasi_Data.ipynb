# **Exploratory Data Analyst (EDA)**

---



**Exploratory Data Analysis (EDA)** adalah proses menganalisis data untuk memahami pola, mendeteksi anomali, dan memeriksa hubungan antar variabel menggunakan teknik statistik dan visualisasi. Tujuannya adalah untuk memperoleh wawasan awal yang membantu dalam mempersiapkan data untuk analisis atau pemodelan lebih lanjut.

**Tujuan EDA (Exploratory Data Analysis)** adalah untuk memahami struktur dan distribusi data, mendeteksi missing values dan outlier, serta menganalisis hubungan antar fitur. EDA juga bertujuan untuk mengeksplorasi pola dalam data dan mempersiapkan data untuk langkah-langkah pra-pemrosesan yang diperlukan sebelum pemodelan, seperti imputasi atau penghapusan data yang tidak relevan.


# **Import library yang dibutuhkan untuk analisis data dan visualisasi**

---



```python
# Import library yang dibutuhkan untuk analisis data dan visualisasi

import pandas as pd
# Fungsi: pandas digunakan untuk memanipulasi dan menganalisis data dalam bentuk tabel (DataFrame).
# Tujuan: Membaca dataset (misalnya file CSV), membersihkan data, mengolah data numerik dan kategorikal.

import seaborn as sns
# Fungsi: seaborn digunakan untuk membuat visualisasi data statistik seperti histogram, boxplot, heatmap, dan countplot.
# Tujuan: Memvisualisasikan distribusi data, hubungan antar variabel, dan membantu dalam analisis eksploratif (EDA).

import matplotlib.pyplot as plt
# Fungsi: matplotlib.pyplot menyediakan fungsi dasar untuk menampilkan grafik seperti judul, label, ukuran, dan layout.
# Tujuan: Menampilkan dan mengatur grafik yang dibuat, baik dari seaborn maupun matplotlib secara langsung.
```

# **Load dataset**

---



```python
# Load dataset
df = pd.read_csv("ObesityDataSet.csv")
# Fungsi: Membaca file CSV bernama "ObesityDataSet.csv" dan menyimpannya ke dalam variabel df (dataframe).
# Tujuan: Mengimpor data dari file eksternal agar dapat dianalisis dan divisualisasikan menggunakan pandas.
```

# **Cek Missing Values**

---



```python
# --- 1. Cek Missing Values ---
print("Missing Values per Kolom:")
print(df.isnull().sum())
# Fungsi: Mengecek jumlah data yang hilang (missing values) pada setiap kolom dalam dataframe.
# Tujuan: Mengidentifikasi kolom mana yang memiliki nilai kosong (NaN), sehingga bisa diputuskan apakah perlu diisi (imputasi) atau dihapus.
```

**Tujuan:**

- Mengetahui Data yang Hilang: Mengecek apakah ada nilai yang hilang (missing values) dalam dataset. Nilai yang hilang dapat memengaruhi analisis atau model yang akan dibangun.

- Keputusan Imputasi atau Penghapusan: Setelah mengetahui kolom mana yang memiliki nilai kosong, kita bisa memutuskan apakah akan mengisi nilai tersebut (imputasi) atau menghapus baris atau kolom yang memiliki missing values.

**Penjelasan Kode:**
- df.isnull(): Mengecek setiap sel dalam dataset, menghasilkan nilai True jika data kosong (NaN), dan False jika ada data.

- sum(): Menghitung jumlah nilai True untuk setiap kolom, yang menunjukkan berapa banyak data kosong pada masing-masing kolom.

**Kenapa Ini Penting?:**

- Mengidentifikasi Masalah Data: Missing values sering kali muncul dalam dataset dan bisa memengaruhi hasil analisis atau pemodelan.

- Persiapan Data: Sebelum melanjutkan analisis atau membangun model, kita perlu memastikan bahwa tidak ada missing values yang dapat mengganggu proses tersebut.

 **Langkah Selanjutnya:**

- Imputasi atau Penghapusan: Berdasarkan hasil cek missing values, kita bisa memilih apakah akan mengisi nilai kosong dengan rata-rata, median, atau metode lainnya, atau menghapus baris atau kolom yang memiliki missing values.

# **Konversi Kolom Numerik**

---



```python
# --- 2. Konversi Kolom Numerik ---
numerical_cols = ["Age", "Height", "Weight", "FCVC", "NCP", "FAF", "TUE", "CH2O"]
df[numerical_cols] = df[numerical_cols].apply(pd.to_numeric, errors='coerce')
# Fungsi:
# - Mendefinisikan daftar kolom yang seharusnya berisi data numerik.
# - Mengonversi nilai pada kolom tersebut ke tipe data numerik (float/int).
# - 'errors="coerce"' akan mengubah nilai yang tidak bisa dikonversi menjadi NaN (missing).
# Tujuan:
# - Memastikan bahwa kolom-kolom tersebut dapat digunakan untuk analisis statistik dan visualisasi numerik.
# - Menangani data yang mungkin salah input atau memiliki format string.
```

**Tujuan:**

- Pastikan Data Sesuai: Mengubah kolom yang seharusnya berisi angka menjadi tipe data angka (misalnya, float atau int).

- Tangani Kesalahan Data: Jika ada data yang tidak bisa dikonversi (misalnya teks dalam kolom angka), akan diganti dengan NaN (nilai kosong).

**Bagaimana Cara Kerjanya?:**

- Daftar Kolom Numerik:
numerical_cols = ["Age", "Height", "Weight", "FCVC", "NCP", "FAF", "TUE", "CH2O"]
Kolom-kolom ini seharusnya berisi angka (misalnya umur, tinggi badan, berat badan, dll).

- Konversi ke Angka:
df[numerical_cols] = df[numerical_cols].apply(pd.to_numeric, errors='coerce')
Kita mengonversi semua nilai dalam kolom-kolom tersebut menjadi angka. Kalau ada data yang tidak bisa dikonversi (misalnya teks), nilai tersebut menjadi NaN.

**Kenapa Ini Penting?:**

- Data yang Konsisten: Pastikan kolom-kolom numerik berisi angka, supaya kita bisa menganalisisnya dengan benar.

- Mempermudah Analisis: Setelah dikonversi, kita bisa menghitung rata-rata, membuat grafik, dan melakukan analisis statistik lainnya.

- Tangani Kesalahan Input: Jika ada kesalahan data, kita bisa segera tahu karena akan muncul sebagai NaN.

**Langkah Selanjutnya:**

- Periksa Nilai NaN: Setelah konversi, cek apakah ada nilai yang hilang dan tentukan apakah perlu diisi atau dihapus.



# **Statistik Deskriptif**

---





```python
# --- 3. Statistik Deskriptif ---
print("\nStatistik Deskriptif:")
print(df[numerical_cols].describe())
# Fungsi:
# - Menampilkan ringkasan statistik deskriptif dari kolom-kolom numerik.
# - Fungsi describe() menghasilkan: count, mean, std, min, 25%, 50% (median), 75%, dan max.
# Tujuan:
# - Memberikan gambaran umum tentang sebaran data (contoh: rata-rata berat badan, rentang usia).
# - Membantu mendeteksi outlier atau nilai yang tidak wajar sejak awal eksplorasi data.
```

**Tujuan:**

- Menampilkan Gambaran Data: Dengan menggunakan describe(), kita mendapatkan ringkasan statistik dari kolom-kolom numerik, seperti rata-rata, nilai minimum, maksimum, dan lainnya.

- Identifikasi Pola dan Masalah: Statistik deskriptif membantu kita melihat pola data (misalnya rata-rata usia atau berat badan) dan mendeteksi masalah seperti outlier.

**Penjelasan Kode:**
- df[numerical_cols]: Mengambil kolom-kolom yang berisi data numerik.
- describe(): Fungsi ini menghasilkan statistik deskriptif untuk kolom-kolom tersebut, seperti:
- count: Jumlah data yang valid (tidak ada NaN).
- mean: Rata-rata nilai.
- std: Standar deviasi (sebaran nilai).
- min dan max: Nilai minimum dan maksimum.
- 25%, 50% (median), 75%: Kuartil data.

**Kenapa Ini Penting?:**

- Pahami Data Lebih Dalam: Memberikan gambaran tentang sebaran data, seperti nilai rata-rata atau rentang nilai (misalnya, rentang usia atau berat badan).

- Deteksi Outlier: Menyediakan informasi yang membantu kita mendeteksi nilai-nilai yang tidak wajar (misalnya, usia yang sangat rendah atau tinggi).

- Bantu Pengambilan Keputusan: Statistik deskriptif adalah langkah awal yang baik sebelum melakukan analisis lebih lanjut atau membuat model.

**Langkah Lanjutan:**

- Analisis Lebih Lanjut: Setelah melihat statistik deskriptif, kita bisa melakukan analisis lebih dalam, seperti menangani outlier atau melihat distribusi data.



# **Visualisasi Distribusi Numerik**

---







```python
# --- 4. Visualisasi Distribusi Numerik ---
for col in numerical_cols:
    plt.figure(figsize=(6, 4))  # Mengatur ukuran figure (lebar x tinggi)
    sns.histplot(df[col].dropna(), kde=True, bins=30)  # Membuat histogram dengan garis KDE
    plt.title(f'Distribusi {col}')  # Judul grafik berdasarkan nama kolom
    plt.xlabel(col)  # Label sumbu X
    plt.ylabel('Frekuensi')  # Label sumbu Y
    plt.tight_layout()  # Mengatur tata letak agar tidak tumpang tindih
    plt.show()  # Menampilkan grafik

# Fungsi:
# - Melakukan iterasi pada setiap kolom numerik dan memvisualisasikan distribusinya menggunakan histogram.
# - Menambahkan kurva KDE (Kernel Density Estimation) untuk menunjukkan bentuk sebaran data secara halus.

# Tujuan:
# - Melihat pola distribusi data: normal, miring (skewed), bimodal, dll.
# - Membantu mendeteksi outlier atau ketidakwajaran dalam data numerik.
```

**Tujuan:**

- Melihat Pola Distribusi Data: Menggunakan histogram dan kurva KDE untuk melihat bagaimana data tersebar. Ini membantu kita untuk mengenali apakah data mengikuti distribusi normal, miring, atau bahkan bimodal.

- Mendeteksi Outlier: Visualisasi ini juga berguna untuk melihat apakah ada nilai yang sangat berbeda dari data lainnya (outlier).

**Penjelasan Kode:**
- for col in numerical_cols: Looping untuk setiap kolom numerik yang ingin divisualisasikan.

- sns.histplot(): Membuat histogram dengan tambahan kurva KDE (Kernel Density Estimation) yang memberikan gambaran lebih halus tentang distribusi data.

- plt.show(): Menampilkan grafik yang telah dibuat untuk setiap kolom numerik.

**Kenapa Ini Penting?:**

- Menganalisis Sebaran Data: Membantu kita untuk memahami distribusi data, apakah data tersebut normal, miring (skewed), atau memiliki lebih dari satu puncak (bimodal).

- Deteksi Outlier: Menampilkan grafik memungkinkan kita untuk melihat apakah ada nilai yang sangat tinggi atau rendah (outlier), yang bisa mempengaruhi analisis lebih lanjut.

- Persiapan untuk Model: Memahami distribusi data membantu kita dalam memilih metode analisis atau model yang tepat (misalnya, transformasi data jika data sangat miring).

**Langkah Selanjutnya:**

- Tindakan Lanjutan: Setelah visualisasi, jika ditemukan outlier atau distribusi yang tidak normal, kita bisa melakukan transformasi data atau menangani outlier sesuai kebutuhan.

# **Kesimpulan Output Visualisasi Distribusi Data Numerik**
**Pola Distribusi Data Terlihat Jelas:**

Melalui histogram dan kurva KDE, kita dapat melihat bentuk distribusi dari setiap kolom numerik. Apakah data terdistribusi secara normal (simetris), miring ke kanan (positif), miring ke kiri (negatif), atau bahkan bimodal (dua puncak distribusi).

**Identifikasi Outlier:**

Nilai-nilai ekstrem yang berada jauh dari puncak utama distribusi mudah dikenali. Ini membantu dalam proses deteksi outlier yang mungkin perlu ditangani dalam tahap preprocessing.

**Pemahaman Skewness (Kemiringan Distribusi):**

Kurva KDE membantu mengidentifikasi apakah distribusi condong ke satu sisi. Ini penting untuk menentukan apakah diperlukan transformasi data (misal: log-transform) untuk analisis statistik atau pemodelan machine learning.

**Perbandingan Antar Fitur Numerik:**

Dengan visualisasi tiap kolom, kita bisa membandingkan seberapa tersebar data antar fitur numerik. Misalnya, fitur dengan persebaran yang sangat lebar atau yang cenderung memiliki konsentrasi nilai pada titik tertentu.

**Kondisi Data yang Lengkap (tanpa NaN):**

Karena dropna() digunakan, visualisasi hanya menampilkan data yang valid (tidak kosong), memastikan interpretasi distribusi tidak bias akibat missing values.

# **Cara Membaca Output Visualisasi Distribusi Numerik**
**1. Histogram (Batang-batang vertikal)**
Menunjukkan frekuensi (jumlah) data dalam rentang tertentu.

- Sumbu X = nilai-nilai dalam kolom numerik.

- Sumbu Y = berapa kali nilai dalam rentang itu muncul (frekuensi).

**Interpretasi:**

Jika batang paling tinggi berada di tengah dan simetris, maka distribusi kemungkinan normal.

Jika lebih banyak batang ke kiri atau kanan → distribusi skewed (miring).

**2. Kurva KDE (Garis lengkung)**
- Memberikan visualisasi bentuk sebaran data yang lebih halus dibanding histogram.

- Membantu melihat apakah distribusi:

- Normal → bentuk lonceng (bell-shaped).

- Skewed ke kanan → ekor ke kanan lebih panjang.

- Skewed ke kiri → ekor ke kiri lebih panjang.

- Bimodal → terdapat dua puncak → indikasi bahwa ada dua kelompok data berbeda.

**3. Outlier**

- Terlihat dari batang histogram atau kurva KDE yang terpencil jauh dari pusat distribusi.

- Perlu diperhatikan apakah outlier wajar atau kesalahan data.
 **Contoh Pembacaan:**

Misalkan kamu melihat grafik untuk kolom age:
-Histogram membentuk lonceng simetris → distribusi normal.

- Kurva KDE mengikuti bentuk histogram → tidak banyak outlier.

- Jika kurva KDE tinggi di satu sisi lalu melandai perlahan ke sisi lain → distribusi miring (skewed).

# **Korelasi Fitur Numerik**

---



```python
# --- 6. Korelasi Fitur Numerik ---
plt.figure(figsize=(10, 8))  # Mengatur ukuran grafik heatmap
corr = df[numerical_cols].corr()  # Menghitung matriks korelasi antar fitur numerik
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
# Membuat heatmap dari korelasi, dengan anotasi angka dan skema warna coolwarm

plt.title("Korelasi antara Fitur Numerik")  # Judul grafik
plt.show()  # Menampilkan grafik

# Fungsi:
# - Menghitung korelasi Pearson antar fitur numerik (nilai antara -1 hingga 1).
# - Menampilkan hasil korelasi dalam bentuk heatmap agar lebih mudah dibaca secara visual.

# Tujuan:
# - Mengetahui hubungan antar fitur numerik, misalnya apakah berat badan berkorelasi kuat dengan tinggi badan.
# - Membantu dalam pemilihan fitur (feature selection) dan memahami multikolinearitas antar variabel.
```

**Tujuan:**

- Melihat Hubungan Antar Fitur: Menggunakan heatmap untuk menampilkan korelasi antar fitur numerik, membantu kita memahami bagaimana fitur-fitur tersebut saling berhubungan.

- Penting untuk Pemilihan Fitur: Memahami korelasi antar fitur sangat penting untuk memilih fitur yang relevan dan menghindari multikolinearitas dalam model.

**Penjelasan Kode:**
- df[numerical_cols].corr(): Menghitung matriks korelasi Pearson antara kolom-kolom numerik.

- sns.heatmap(): Menampilkan matriks korelasi dalam bentuk heatmap, yang memudahkan visualisasi hubungan antar fitur numerik. Angka-angka di dalam heatmap menunjukkan kekuatan korelasi antara fitur (nilai antara -1 hingga 1).

- cmap='coolwarm': Skema warna yang menunjukkan korelasi positif dengan warna lebih hangat dan korelasi negatif dengan warna lebih dingin.

**Kenapa Ini Penting?:**

- Memahami Hubungan Antar Fitur: Mengetahui fitur mana yang saling berhubungan sangat berguna dalam analisis data dan pemodelan. Misalnya, jika dua fitur sangat berkorelasi, kita bisa memilih salah satu untuk mengurangi redundansi.

- Menghindari Multikolinearitas: Jika fitur-fitur sangat berkorelasi, model bisa menjadi kurang stabil dan lebih rentan terhadap overfitting. Ini dapat membantu kita memilih fitur yang benar-benar relevan untuk model.

**Langkah Selanjutnya:**

- Pilih Fitur yang Relevan: Berdasarkan korelasi yang ditemukan, kita bisa memilih fitur-fitur yang paling penting dan mengurangi fitur yang sangat berkorelasi.

- Periksa Multikolinearitas: Jika ada korelasi tinggi antara fitur, pertimbangkan untuk menghilangkan salah satu fitur tersebut atau menggunakan teknik seperti PCA (Principal Component Analysis).

# **Kesimpulan Output Heatmap Korelasi Fitur Numerik**
**Korelasi Antar Fitur Terlihat Jelas:**

Setiap sel pada heatmap menunjukkan nilai korelasi Pearson antara dua fitur numerik, dengan nilai berkisar dari -1 hingga 1.

**Interpretasi Nilai Korelasi:**

- 1.00 → Korelasi positif sempurna (fitur meningkat bersama).

- -1.00 → Korelasi negatif sempurna (satu naik, yang lain turun).

- 0.00 → Tidak ada korelasi linear.

**Mendeteksi Hubungan Kuat:**

- Korelasi mendekati 1 atau -1 menandakan hubungan linear kuat.

- Korelasi mendekati 0 berarti hubungan lemah atau tidak linear.

**Warna Membantu Pembacaan Cepat:**

**Skema warna coolwarm digunakan:**

- Biru → Korelasi negatif.

- Merah → Korelasi positif.

- Putih/pucat → Korelasi mendekati nol.

**Manfaat untuk Feature Selection:**

- Jika dua fitur sangat berkorelasi (misalnya 0.9 atau lebih), mungkin salah satunya bisa dihapus untuk menghindari multikolinearitas.

# **Cara Membaca Heatmap Korelasi**
**Titik Perpotongan Baris dan Kolom:**

- Menunjukkan korelasi antara dua fitur numerik, misalnya antara Weight dan BMI.

**Warna & Angka:**

- Warna semakin merah → korelasi positif kuat.

- Warna semakin biru → korelasi negatif kuat.

- Angka di setiap kotak adalah nilai korelasinya (dengan 2 desimal).

**Contoh Interpretasi:**
Jika kotak antara Weight dan BMI menunjukkan nilai 0.92 (merah) → berarti korelasi positif sangat kuat.

Jika Age dan Height menunjukkan -0.05 (pucat) → tidak ada korelasi signifikan.

 **Mengapa Ini Penting?**
- Menghindari Redundansi: Dua fitur yang sangat berkorelasi bisa menyumbang informasi yang sama.

- Meningkatkan Kinerja Model: Dengan menghapus fitur yang redundant, model bisa lebih sederhana dan akurat.

- Mendeteksi Struktur Data: Heatmap membantu memahami bagaimana fitur saling berkaitan secara numerik.
